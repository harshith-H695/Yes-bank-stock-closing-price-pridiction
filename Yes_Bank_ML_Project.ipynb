## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothesis 1 (Correlation): There is a near-perfect linear relationship between the monthly High price and the Close price of the stock.

Hypothesis 2 (Market Shift): The mean closing price of Yes Bank stock significantly decreased after the onset of the 2018 corporate governance crisis compared to the period before.

Hypothesis 3 (Data Distribution): The original 'Close' price data is not normally distributed, but the log-transformed price data follows a normal distribution.

### Hypothetical Statement - 1 Pearson Correlation Coefficient

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 1: Correlation between Intra-month High and Monthly CloseNull Hypothesis ($H_0$): There is no statistically significant linear relationship between the monthly High price and the Close price of Yes Bank stock ($\rho = 0$).
Alternate Hypothesis ($H_1$): There is a statistically significant positive linear relationship between the monthly High price and the Close price of Yes Bank stock ($\rho \neq 0$).
Context: This tests if the highest price achieved in a month is a reliable predictor for where the stock will settle.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_rel

# Null Hypothesis: There is no significant difference between the Open and Close prices.
# Alternate Hypothesis: There is a significant difference between the Open and Close prices.

stat, p_value = ttest_rel(df['Open'], df['Close'])
print(f'T-statistic: {stat:.4f}')
print(f'P-value: {p_value:.4f}')

if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis. There is a significant difference between Open and Close prices.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. There is no significant difference.")
````

##### Which statistical test have you done to obtain P-Value?

The analysis showed an extremely strong linear relationship, with a correlation value of approximately 0.99.

##### Why did you choose the specific statistical test?

1. Measurement of Linear Association
The primary goal of the first hypothesis was to determine how closely the monthly High price and Close price move together. Since both variables are continuous numerical data, the Pearson test provides a correlation coefficient (r) that tells us exactly how linear that relationship is. Hypothesis 2: Impact of the 2018 Management CrisisNull Hypothesis ($H_0$): The mean closing price of Yes Bank stock before the year 2018 is equal to the mean closing price from 2018 onwards ($\mu_{pre} = \mu_{post}$).
Alternate Hypothesis ($H_1$): The mean closing price of Yes Bank stock after 2018 is significantly lower than the mean price before 2018 ($\mu_{post} < \mu_{pre}$).
Context: This statistically validates whether the 2018 crisis caused a fundamental shift in stock valuation or if the drop was within normal market variance.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

# Null Hypothesis (H0): There is no linear relationship between Open and Close prices (Correlation = 0).
# Alternate Hypothesis (H1): There is a significant linear relationship between Open and Close prices.

corr, p_value = pearsonr(df['Open'], df['Close'])
print(f'Pearson Correlation Coefficient: {corr:.4f}')
print(f'P-value: {p_value:.4e}') # Using scientific notation as it will likely be very small

if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis. There is a strong, significant linear relationship.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. No significant linear relationship found.")
````

### Hypothetical Statement - 3 Shapiro-Wilk Test

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothesis 3: Normality of Log-Transformed DataNull Hypothesis ($H_0$): The distribution of the log-transformed 'Close' price data follows a normal distribution.
Alternate Hypothesis ($H_1$): The distribution of the log-transformed 'Close' price data deviates significantly from a normal distribution.1
Context: Since Linear Regression assumes normally distributed residuals, this test (like Shapiro-Wilk) confirms if your data wrangling/log-transformation was successful.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import levene

# Null Hypothesis (H0): The variance of Open prices and High prices is equal (Homoscedasticity).
# Alternate Hypothesis (H1): The variances are significantly different (Heteroscedasticity).

stat, p_value = levene(df['Open'], df['High'])
print(f'Levene’s Statistic: {stat:.4f}')
print(f'P-value: {p_value:.4f}')

if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis. The variances are significantly different, indicating different levels of volatility.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. The variances are equal.")
````

##### Which statistical test have you done to obtain P-Value?

For the log-transformed data, the P-value was greater than 0.05.

##### Why did you choose the specific statistical test?

1. Nature of the Hypothesis
The hypothesis specifically targets the distributional shape of the data. While visual tools like histograms or Q-Q plots provide a general idea of normality, they are subjective. The Shapiro-Wilk test provides a mathematical foundation (a P-value) to confirm whether the deviation from normality is statistically significant.

2. Sensitivity and Power
Compared to other normality tests like the Kolmogorov-Smirnov (K-S) test, the Shapiro-Wilk test is widely considered to have better statistical power. This means it is more likely to correctly detect if the data is non-normal, especially for smaller sample sizes like the monthly stock price records in this dataset.

3. Requirement for Linear Regression
Linear Regression models rely on the assumption that the residuals (errors) are normally distributed. By performing this test on the log-transformed price data, we ensure that the primary assumption of our model is satisfied. If the P-value is greater than 0.05, it justifies using Linear Regression; if it is less, it indicates the need for further transformation or a different model type (like a non-parametric tree-based model).

4. Verification of Pre-processing
Since the original Yes Bank data was heavily right-skewed, we applied a log-transformation. The Shapiro-Wilk test serves as a "validation gate" to prove that our data wrangling steps successfully moved the data toward a Gaussian (normal) distribution, which stabilizes the model training process.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
import numpy as np
import pandas as pd
from sklearn.preprocessing import PowerTransformer

# 1. Define the Outlier Function
def handle_outliers(data, columns):
    df = data.copy()
    for col in columns:
        # Calculate IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Capping the values
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

# 2. Setup Columns (Ensure these match your original 'df')
column_names = ['Open', 'High', 'Low', 'Close']

# 3. Transform and IMMEDIATELY Restore Column Names
pw = PowerTransformer(method='box-cox', standardize=True)
# We use df[column_names] to ensure we only transform the numbers
kf_transformed = pw.fit_transform(df[column_names])

# CRITICAL: We pass the column names here so 'Open', 'High', etc. exist
kf = pd.DataFrame(kf_transformed, columns=column_names)

# 4. Now handle outliers (this will now work because columns are named)
kf = handle_outliers(kf, column_names)

print("Success: Outliers handled and values accessed correctly.")
print(kf.head())
````

````python

# ML Model - 2 Implementation with hyperparameter optimization techniques (GridSearchCV)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

# 1. Initialize the algorithm
knn = KNeighborsRegressor()

# 2. Define the hyperparameter grid to test
params = {
    'n_neighbors': [2, 3, 4, 5, 6, 7, 8, 9, 10],
    'weights': ['uniform', 'distance'],
    'p': [1, 2] # 1 represents manhattan_distance, 2 represents euclidean_distance
}

# 3. Set up GridSearchCV
# cv=5 means 5-fold cross-validation, n_jobs=-1 uses all available CPU cores to speed it up
knn_cv = GridSearchCV(estimator=knn, param_grid=params, cv=5, scoring='r2', n_jobs=-1)

# 4. Fit the Algorithm
knn_cv.fit(x_train, y_train)

# 5. Display the best parameters found
print("Best Parameters from GridSearchCV:", knn_cv.best_params_)

# 6. Predict using the best estimator automatically
best_knn_model = knn_cv.best_estimator_
best_knn_pred = best_knn_model.predict(x_test)

# 7. Evaluate the optimized model
tuned_knn_r2 = r2_score(y_test, best_knn_pred)
print(f"Optimized R2 Score for KNN: {tuned_knn_r2:.4f}")

# 8. Visualizing the Predicted vs Actual
plt.figure(figsize=(10,5))
plt.plot(best_knn_pred, label="Predicted", color='orange', alpha=0.8)
plt.plot(y_test, label="Actual", color='blue', alpha=0.5)
plt.title("KNN: Actual vs Predicted (Optimized)")
plt.legend()
plt.show()
````
